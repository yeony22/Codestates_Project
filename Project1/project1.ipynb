```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import io
```

```
from google.colab import files
myfile = files.upload()
```

```
data = pd.read_csv(io.BytesIO(myfile['project1.csv'])) #데이터 불러오기
data
```

```
df = data.drop(['Unnamed: 0'], axis=1) #인덱스 정리
```

```
df.isnull().sum()
```

```
df_drop = df.dropna(subset=['Year']) #Year이 null값인 행 삭제
df_drop = df_drop.fillna('missing')
```

```
list(df_drop['Year'])
```

```
year = list(df_drop['Year'])
year = list(map(int, year))

for i in range(len(year)):
  if year[i] < 21:
    year[i] += 2000
  elif year[i] < 100:
    year[i] += 1900

df_drop["Year"] = year #Year 전처리
```

```
list(df_drop['NA_Sales'])
```

```
na_sales = list(df_drop['NA_Sales'])

for i in range(len(na_sales)):
    if '.' in na_sales[i]:
      if 'M' in na_sales[i]:

        na_sales[i] = na_sales[i][:-1] #마지막 문자 없애기
        na_sales[i] = na_sales[i][0:1] + na_sales[i][2:] #점 없애기
        na_sales[i] = na_sales[i] + '000' # 단위 바꾸기

        if na_sales[i][:1] == '0':
          na_sales[i] = na_sales[i][1:] #0 없애기
          if na_sales[i][:1] == '0':
            na_sales[i] = na_sales[i][1:] #0 없애기

    elif '0M' in na_sales[i]:
      na_sales[i] = '0'

    elif 'K' in na_sales[i]:
      na_sales[i] = na_sales[i][:-1] #마지막 문자열 없애기
      if na_sales[i] != '0':
        na_sales[i] = na_sales[i] + '000'

na_sales = list(map(float, na_sales))
df_drop["NA_Sales"] = na_sales #NA_Sales 전처리
```

```
eu_sales = list(df_drop['EU_Sales'])

for i in range(len(eu_sales)):
    if '.' in eu_sales[i]:
      if 'M' in eu_sales[i]:

        eu_sales[i] = eu_sales[i][:-1] #마지막 문자 없애기
        eu_sales[i] = eu_sales[i][0:1] + eu_sales[i][2:] #점 없애기
        eu_sales[i] = eu_sales[i] + '000' # 단위 바꾸기

        if eu_sales[i][:1] == '0':
          eu_sales[i] = eu_sales[i][1:] #0 없애기
          if eu_sales[i][:1] == '0':
            eu_sales[i] = eu_sales[i][1:] #0 없애기

    elif '0M' in eu_sales[i]:
      eu_sales[i] = '0'

    elif 'K' in eu_sales[i]:
      eu_sales[i] = eu_sales[i][:-1] #마지막 문자열 없애기
      if eu_sales[i] != '0':
        eu_sales[i] = eu_sales[i] + '000'

eu_sales = list(map(float, eu_sales))
df_drop["EU_Sales"] = eu_sales #EU_Sales 전처리
```

```
jp_sales = list(df_drop['JP_Sales'])

for i in range(len(jp_sales)):
    if '.' in jp_sales[i]:
      if 'M' in jp_sales[i]:

        jp_sales[i] = jp_sales[i][:-1] #마지막 문자 없애기
        jp_sales[i] = jp_sales[i][0:1] + jp_sales[i][2:] #점 없애기
        jp_sales[i] = jp_sales[i] + '000' # 단위 바꾸기

        if jp_sales[i][:1] == '0':
          jp_sales[i] = jp_sales[i][1:] #0 없애기
          if jp_sales[i][:1] == '0':
            jp_sales[i] = jp_sales[i][1:] #0 없애기

    elif '0M' in jp_sales[i]:
      jp_sales[i] = '0'

    elif 'K' in jp_sales[i]:
      jp_sales[i] = jp_sales[i][:-1] #마지막 문자열 없애기
      if jp_sales[i] != '0':
        jp_sales[i] = jp_sales[i] + '000'

jp_sales = list(map(float, jp_sales))
df_drop["JP_Sales"] = jp_sales #JP_Sales 전처리
```

```
ot_sales = list(df_drop['Other_Sales'])

for i in range(len(ot_sales)):
    if '.' in ot_sales[i]:
      if 'M' in ot_sales[i]:

        ot_sales[i] = ot_sales[i][:-1] #마지막 문자 없애기
        ot_sales[i] = ot_sales[i][0:1] + ot_sales[i][2:] #점 없애기
        ot_sales[i] = ot_sales[i] + '000' # 단위 바꾸기

        if ot_sales[i][:1] == '0':
          ot_sales[i] = ot_sales[i][1:] #0 없애기
          if ot_sales[i][:1] == '0':
            ot_sales[i] = ot_sales[i][1:] #0 없애기

    elif '0M' in ot_sales[i]:
      ot_sales[i] = '0'

    elif 'K' in ot_sales[i]:
      ot_sales[i] = ot_sales[i][:-1] #마지막 문자열 없애기
      if ot_sales[i] != '0':
        ot_sales[i] = ot_sales[i] + '000'

ot_sales = list(map(float, ot_sales))
df_drop["Other_Sales"] = ot_sales #Other_Sales 전처리
```

### 1. 지역에 따라서 선호하는 게임 장르가 다를까

```
Action = df_drop['Genre'] == 'Action'
action = df_drop[Action]

Adventure = df_drop['Genre'] == 'Adventure'
adventure = df_drop[Adventure]

Misc = df_drop['Genre'] == 'Misc'
misc = df_drop[Misc]

Platform = df_drop['Genre'] == "Platform"
platform = df_drop[Platform]

Sports = df_drop['Genre'] == "Sports"
sports = df_drop[Sports]

Simulation = df_drop['Genre'] == "Simulation"
simulation = df_drop[Simulation]

Racing = df_drop['Genre'] == "Racing"
racing = df_drop[Racing]

RolePlaying = df_drop['Genre'] == "Role-Playing"
rolePlaying = df_drop[RolePlaying]

Puzzle = df_drop['Genre'] == "Puzzle"
puzzle = df_drop[Puzzle]

Strategy = df_drop['Genre'] == "Strategy"
strategy = df_drop[Strategy]

Fighting = df_drop['Genre'] == "Fighting"
fighting = df_drop[Fighting]

Shooter = df_drop['Genre'] == "Shooter"
shooter = df_drop[Shooter]
```

```
genre = list(df_drop['Genre'].unique())
genre = genre[:-1]
```

```
na_sum = [action['NA_Sales'].sum(), adventure['NA_Sales'].sum(), misc['NA_Sales'].sum(), platform['NA_Sales'].sum(), sports['NA_Sales'].sum(), simulation['NA_Sales'].sum(), racing['NA_Sales'].sum(), rolePlaying['NA_Sales'].sum(), puzzle['NA_Sales'].sum(), strategy['NA_Sales'].sum(), fighting['NA_Sales'].sum(), shooter['NA_Sales'].sum()]
eu_sum = [action['EU_Sales'].sum(), adventure['EU_Sales'].sum(), misc['EU_Sales'].sum(), platform['EU_Sales'].sum(), sports['EU_Sales'].sum(), simulation['EU_Sales'].sum(), racing['EU_Sales'].sum(), rolePlaying['EU_Sales'].sum(), puzzle['EU_Sales'].sum(), strategy['EU_Sales'].sum(), fighting['EU_Sales'].sum(), shooter['EU_Sales'].sum()]
jp_sum = [action['JP_Sales'].sum(), adventure['JP_Sales'].sum(), misc['JP_Sales'].sum(), platform['JP_Sales'].sum(), sports['JP_Sales'].sum(), simulation['JP_Sales'].sum(), racing['JP_Sales'].sum(), rolePlaying['JP_Sales'].sum(), puzzle['JP_Sales'].sum(), strategy['JP_Sales'].sum(), fighting['JP_Sales'].sum(), shooter['JP_Sales'].sum()]
```

```
df_genre = pd.DataFrame({'Genre' : genre,
                         'NA_Sales' : na_sum,
                         'EU_Sales' : eu_sum,
                         'JP_Sales' : jp_sum
                         })
df_genre
```
